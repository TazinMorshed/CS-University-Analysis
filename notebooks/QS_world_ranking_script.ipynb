```python
from selenium import webdriver
from selenium.webdriver.common.by import By
import time 
webdriver_path = "C:\Program Files (x86)\chromedriver.exe"
url = "https://www.topuniversities.com/university-rankings/university-subject-rankings/2022/engineering-technology"
driver = webdriver.Chrome(webdriver_path)
driver.get(url)
time.sleep(10)
```

```python
row =[]
links =[]

for i in range(50): 
    time.sleep(2)
    driver.execute_script("window.scrollTo(0, 0)") 
    time.sleep(1)
    driver.execute_script("window.scrollTo(0, 2500)") 

    time.sleep(2)

    rows = driver.find_element(By.ID,"ranking-data-load")
    row_ind = rows.find_elements(By.XPATH,"//div[contains(@class,'row') and contains(@class , 'ind ')]")
    for x in row_ind:
        row.append(x.text)
        links.append(x.find_element(By.TAG_NAME,'a').get_attribute('href'))
    
    next_button  = driver.find_element(By.XPATH,'//a[contains(@class,"page-link") and contains(@class,"next")]')

    next_button.click()
    time.sleep(2)
```

```python
QS_data = []
for i in range(len(links)): 
    driver.get(links[i])

    driver.execute_script("window.scrollTo(0, 2500)") 

    time.sleep(5)
    
    try:
        general_admission = driver.find_element('xpath','//*[@id="admissionTab"]/div[1]').text
        g_admission = general_admission.split('\n')

        bachelor_admission = driver.find_element('xpath','//*[@id="admissionTab"]/div[2]').text
        b_admission = bachelor_admission.split('\n')

        admission_dict = generate_admission_requirement_2(g_admission,b_admission)

    
    except: 

        try:
            general_admission = driver.find_element('xpath','//*[@id="admissionTab"]/div[1]').text
            g_admission = general_admission.split('\n')
            admission_dict = generate_admission_requirement_1(g_admission)
        except:
            admission_dict = generate_admission_requirement_0()


        
    ranking_criteria = driver.find_element(By.CLASS_NAME,"criteria-wrap").text
    ranking_criteria = ranking_criteria.split('\n')

    overall, a_rep, e_rep, faculty_student_ratio, cite_per_faculty, int_faculty_ration, int_student_ratio, int_research_network, e_outcomes = generate_ranking_criteria(ranking_criteria)  
    
    
    QCdata = {}
    QCdata['Rank'] = rank[i] 
    QCdata['University Name'] = university_name[i] 
    QCdata['State']=state[i] 
    QCdata['Country']=country[i] 
    QCdata['Impact factor']=impact_factor[i] 
    
    QCdata.update(admission_dict)
    
    QCdata['Ranking Criteria'] = {}
    QCdata['Ranking Criteria']['Overall'] = overall
    QCdata['Ranking Criteria']['Academic Reputation'] = a_rep
    QCdata['Ranking Criteria']['Employer Reputation'] = e_rep
    QCdata['Ranking Criteria']['Faculty Student Ratio'] = faculty_student_ratio
    QCdata['Ranking Criteria']['Citations per Faculty'] = cite_per_faculty
    QCdata['Ranking Criteria']['International Faculty Ratio'] = int_faculty_ration
    QCdata['Ranking Criteria']['International Students Ratio'] = int_student_ratio
    QCdata['Ranking Criteria']['International Research Network'] = int_research_network
    QCdata['Ranking Criteria']['Employment Outcomes'] = e_outcomes
    
    
    QS_data.append(QCdata)
```
